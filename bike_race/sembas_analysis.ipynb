```python
import matplotlib.pyplot as plt 
import numpy as np 
import json

from numpy import ndarray
from matplotlib.axes import Axes

BOUNDARY_PATH = r"" # TODO, just provide the direct path to SEMBAS/.results folder
BOUNDARY_FILE_NAME = "boundary.json"

"""
The boundary is broken up into two arrays. The first are the positions of the boundary points, the
second is the direction of the surface at each of those positions.
"""
def load_boundary():
    with open(f"{BOUNDARY_PATH}/{BOUNDARY_FILE_NAME}") as f: 
        data = json.load(f)
    return np.array(data["boundary_points"]), np.array(data["boundary_surface"])

def num_dims(bps: np.ndarray) -> bool:
    return len(bps[0])

def is_displayable(bps: np.ndarray) -> bool:
    return num_dims(bps) <= 3
```

```python
bps, osvs = load_boundary()
is_3d = num_dims(bps) == 3
if not is_displayable(bps):
    print("WARNING: Too many dimensions to plot!")
```

```python
SHOW_SURFACE_VECS = True
OSV_SCALE = 0.1

if is_3d:
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
else:
    fig, ax = plt.subplots()

fig.tight_layout()

ax.scatter(*bps.T)

if SHOW_SURFACE_VECS:
    ax.quiver(
        *bps.T,
        *(OSV_SCALE * osvs.T)
    )
    
plt.show()
```
