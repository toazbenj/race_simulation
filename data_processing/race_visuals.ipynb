Parsing Feather Files

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = 'version2_data/weighted_sum/cost_stats.feather'
df = pd.read_feather(path)
df['action1_clean'] = pd.to_numeric(df['action1'], errors='coerce')

# Create bins that align with integer values
min_val = int(df['action1_clean'].min())
max_val = int(df['action1_clean'].max())
bins = np.arange(min_val, max_val + 2) - 0.5  # Bins centered on integers

df['action1_clean'].hist(bins=bins, grid=False, color='steelblue', density=True, align='mid', edgecolor='none')

# Remove border around graph area
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
# plt.gca().spines['bottom'].set_visible(False)
# plt.gca().spines['left'].set_visible(False)

# Set x-axis ticks to show integer values
plt.xticks(range(min_val, max_val))

# Add titles to axes
plt.xlabel('Action 1 Values')
plt.ylabel('Density')
plt.title('Weighted Sum Defender vs Weighted Sum Attacker')

plt.show()
```
