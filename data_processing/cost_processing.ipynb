```python
import matplotlib.pyplot as plt
from pathlib import Path

# extra code – code to save the figures as high-res PNGs for the book

IMAGES_PATH = Path() / "images"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)

# extra code – the next 5 lines define the default font sizes
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

axes = df[["A1_1_1", "B1_2_4", "C1_6_7", "A2_1_1", "B2_2_4", "C2_6_7",
           "Theta_a1", "action1", "action2", "State1_x", "State1_y",
           "State1_v", "State1_phi", "State1_b", "State2_x", "State2_y",
           "State2_v", "State2_phi", "State2_b"]].hist(bins=50, figsize=(12, 8))

for ax in axes.ravel():
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")

save_fig("attribute_histogram_plots")
plt.show()
```

```python
import numpy as np

cleaned_data = df.drop(['action2', 'Theta_a2', 'Theta_b2', 'Theta_c2', 'Action_space1', 'Action_space2', "State1_x", "State1_y",
                     "State1_v", "State1_phi", "State1_b", "State2_x", "State2_y",
                     "State2_v", "State2_phi", "State2_b"], axis=1)

cleaned_data = cleaned_data.drop(cleaned_data.filter(regex='^A2_').columns, axis=1)
cleaned_data = cleaned_data.drop(cleaned_data.filter(regex='^B2_').columns, axis=1)
cleaned_data = cleaned_data.drop(cleaned_data.filter(regex='^C2_').columns, axis=1)


def shuffle_and_split_data(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_set, test_set = shuffle_and_split_data(cleaned_data, 0.2)
print(len(train_set))
print(len(test_set))
```
